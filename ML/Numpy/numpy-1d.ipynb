```python
#Softmax
def stable_softmax(x):
    exps = np.exp(x-x.max())
    return exps/np.sum(exps)
x = np.array([1,2,3])
f = stable_softmax(x)
print(f)
```

Temperature

```python
import pandas as pd
import numpy as np

data = pd.read_csv('temperature-1d.csv').to_numpy()
temp = data[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.plot(temp)
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.show
```

```python
temp_rs = temp.reshape(-1, 24)
print(temp_rs.shape)
```

```python
mean_24 = np.mean(temp_rs, axis=1)
```

```python
print(data.shape)
print(temp.shape)
result = np.repeat(mean_24, 24)
print(result.shape)
```

```python
plt.plot(temp, label='Original Temperature')
plt.plot(result, label='Mean Temperature')
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.title('Temp')
plt.legend()
plt.show()
```

```python
#One hot Encoding
K = 3 #num_labels
N = 9 #num_samples

labels = [0,0,0,1,1,1,2,2,2]
labels = np.array(labels)
print(labels)
```

```python
labels_onehot = np.zeros((N,K))

labels_onehot[np.arange(N), labels] = 1
print(labels_onehot)
```

```python
labels = np.genfromtxt('iris.csv',
                       delimiter=',',
                       usecols=4,
                       dtype='str',
                       skip_header=1)
print(labels.shape)
print(labels[:10])
```

```python
categories = np.unique(labels)
print(categories)

for i in range(3):
    labels = np.char.replace(labels, categories[i], str(i))

labels = labels.astype('float')
print(labels)
```
